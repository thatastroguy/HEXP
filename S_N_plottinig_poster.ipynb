```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
df = pd.read_csv('flux_curves_finerbin.csv', sep=',')
df
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.scatter(df['Sep'], df['F8-24AGN1'], marker='x', color='royalblue')
ax.scatter(df['Sep'], df['F8-24AGN2'], marker='x', color='tab:orange')


ax.hlines((4.274100e-13), -1.5, 60, linestyles='dashed')

ax.set(xlabel="Angular Separation (\'\')", ylabel="Flux 8-24 keV")

#ax.set_ylim(10**-13,2*10**-13)

ax.set_xlim(-1.5,35)


ax.set_yscale('log')

plt.tight_layout()
plt.savefig('hexp_8-24_flux_v_angsep.png', dpi=150)

plt.show()

```
