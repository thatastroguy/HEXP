```python
def plot_specmod(name,fitstat,dof,pars):
    Plot.device = '/null'
    
    #Assigning plot group for data set #1
    Plot('ufspec')
    (energies1, energies1err) = Plot.x(plotGroup=1), Plot.xErr(plotGroup=1)
    (flux1, flux1err) = Plot.y(plotGroup=1), Plot.yErr(plotGroup=1)
    folded1 = Plot.model(plotGroup=1)
    
    #Assigning plot group for data set #1
    Plot('rat')
    (rat1, rat1err) = Plot.x(plotGroup=1), Plot.xErr(plotGroup=1)
    (fluxr1, fluxr1err) = Plot.y(plotGroup=1), Plot.yErr(plotGroup=1)
    
    # Plot using Matplotlib:
    fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]}, figsize=(6,6))
    plt.rcParams.update({'font.size': 12})
    
    #Plotting the data (errors bars and open circles for each):
    im = ax1.errorbar(energies1, flux1, yerr=flux1err, xerr=energies1err, c='royalblue', fmt=',', alpha=0.8, label='LEM')
    im = ax1.scatter(energies1, flux1, alpha=0.6, marker='o', facecolors='none', edgecolors='royalblue', label='_nolegend_')
    
    #Plotting the models:
    im1 = ax1.plot(energies1, folded1, c='black', linestyle = '-', alpha=0.8, label='_nolegend_')    
    
    im = ax2.errorbar(rat1, fluxr1, yerr=fluxr1err, xerr=rat1err, c='royalblue', fmt=',', alpha=0.8, label='FPMA')
    im = ax2.scatter(rat1, fluxr1, alpha=0.6, marker='o', facecolors='none', edgecolors='royalblue', label='_nolegend_')
    
    ax2.set_xscale('log')
    ax2.set(xlabel='Energy (keV)', ylabel='Ratios')
    ax1.set(xlabel='Energy (keV)', ylabel='Normalized counts s$^{-1}$ keV$^{-1}$')
    ax1.set_yscale('log')
    ax1.set_xscale('log')
    xmin, xmax = 0.5, 2.0
    ax1.set_xlim(1.0, 1.8)
    ax2.set_xlim(1.0, 1.8)
    ax2.hlines(1.0, xmin, xmax, 'black', linestyles='dotted') # Plotting a zero line here...
        
    # Now providing fit parameters on our plots
    #ax1.text(0.5, 0.90, 'C-stat='+str(np.round(fitstat, decimals=2)), transform=ax1.transAxes)
    #ax1.text(0.5, 0.82, 'DoF='+str(dof), transform=ax1.transAxes)
    #ax1.text(0.1, 0.90, str(model), transform=ax1.transAxes)
    #ax1.text(0.1, 0.82, 'Free par:'+str(pars), transform=ax1.transAxes)
    
    ax1.legend(loc='upper right', prop={'size': 10})
    plt.tight_layout()
    plt.savefig(str(name[:-4:1])+'model.pdf', dpi=150, overwrite=True)

def plot_bestfit(name,fitstat,dof,pars):
    Plot.device = '/null'
    
    #Assigning plot group for data set #1
    Plot('ufspec')
    (energies1, energies1err) = Plot.x(plotGroup=1), Plot.xErr(plotGroup=1)
    (flux1, flux1err) = Plot.y(plotGroup=1), Plot.yErr(plotGroup=1)
    folded1 = Plot.model(plotGroup=1)
    
    #Assigning plot group for data set #1
    Plot('rat')
    (rat1, rat1err) = Plot.x(plotGroup=1), Plot.xErr(plotGroup=1)
    (fluxr1, fluxr1err) = Plot.y(plotGroup=1), Plot.yErr(plotGroup=1)
    
    # Plot using Matplotlib:
    fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]}, figsize=(6,6))
    plt.rcParams.update({'font.size': 12})
    
    #Plotting the data (errors bars and open circles for each):
    im = ax1.errorbar(energies1, flux1, yerr=flux1err, xerr=energies1err, c='royalblue', fmt=',', alpha=0.8, label='LEM')
    im = ax1.scatter(energies1, flux1, alpha=0.6, marker='o', facecolors='none', edgecolors='royalblue', label='_nolegend_')
    
    #Plotting the models:
    im1 = ax1.plot(energies1, folded1, c='black', linestyle = '-', alpha=0.8, label='_nolegend_')    
    
    im = ax2.errorbar(rat1, fluxr1, yerr=fluxr1err, xerr=rat1err, c='royalblue', fmt=',', alpha=0.8, label='FPMA')
    im = ax2.scatter(rat1, fluxr1, alpha=0.6, marker='o', facecolors='none', edgecolors='royalblue', label='_nolegend_')
    
    ax2.set_xscale('log')
    ax2.set(xlabel='Energy (keV)', ylabel='Ratios')
    ax1.set(xlabel='Energy (keV)', ylabel='Normalized counts s$^{-1}$ keV$^{-1}$')
    ax1.set_yscale('log')
    ax1.set_xscale('log')
    xmin, xmax = 0.5, 2.0
    ax1.set_xlim(1.0, 1.8)
    ax2.set_xlim(1.0, 1.8)
    ax2.set_ylim(-1.0,5.0)
    ax2.hlines(1.0, xmin, xmax, 'black', linestyles='dotted') # Plotting a zero line here...
        
    # Now providing fit parameters on our plots
    #ax1.text(0.5, 0.90, 'C-stat='+str(np.round(fitstat, decimals=2)), transform=ax1.transAxes)
    #ax1.text(0.5, 0.82, 'DoF='+str(dof), transform=ax1.transAxes)
    #ax1.text(0.1, 0.90, str(model), transform=ax1.transAxes)
    #ax1.text(0.1, 0.82, 'Free par:'+str(pars), transform=ax1.transAxes)
    
    ax1.legend(loc='upper right', prop={'size': 10})
    plt.tight_layout()
    plt.savefig(str(name[:-4:1])+'best_fit.pdf', dpi=150, overwrite=True)


```
