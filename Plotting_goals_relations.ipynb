```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM # importing cosmology package from astropy
cosmo = FlatLambdaCDM(H0=70, Om0=0.3) # assigning cosmology
from astropy import units as u


```

```python
# read in the csv file as a pandas DataFrame
df = (pd.read_csv('Table1_lum_NH.dat', sep=',')).dropna(subset=['NH'])

# create an empty column called 'flag'
df['L10-24obs_flag'] = '' #L10-24obs	L2-10intr	L10-24intr	

# iterate over each row in the DataFrame
for index, row in df.iterrows():

    # check if the special character is present in the column
    if str(row['L10-24obs'])[0] in ['<', '>']:

        # if the special character is present, add it to the 'flag' column
        df.loc[index, 'L10-24obs_flag'] = str(row['L10-24obs'])[0]

        # remove the special character from the original column
        df.loc[index, 'L10-24obs'] = float(str(row['L10-24obs'])[1:])

    # check if the special character is present in the column
    if str(row['L2-10intr'])[0] in ['<', '>']:

        # if the special character is present, add it to the 'flag' column
        df.loc[index, 'L2-10intr_flag'] = str(row['L2-10intr'])[0]

        # remove the special character from the original column
        df.loc[index, 'L2-10intr'] = float(str(row['L2-10intr'])[1:])

    # check if the special character is present in the column
    if str(row['L10-24intr'])[0] in ['<', '>']:

        # if the special character is present, add it to the 'flag' column
        df.loc[index, 'L10-24intr_flag'] = str(row['L10-24intr'])[0]

        # remove the special character from the original column
        df.loc[index, 'L10-24intr'] = float(str(row['L10-24intr'])[1:])

    # check if the special character is present in the column
    if str(row['NH'])[0] in ['<', '>']:

        # if the special character is present, add it to the 'flag' column
        df.loc[index, 'NH_flag'] = str(row['NH'])[0]

        # remove the special character from the original column
        df.loc[index, 'NH'] = float(str(row['NH'])[1:])

# write the modified DataFrame to a new csv file
df.to_csv('output.csv', sep=',', index=False)
```

```python
df = pd.read_csv('output.csv', sep=',')
```

```python
df
```

```python
df['DL(MPC)'] = cosmo.luminosity_distance(df['z'])/u.Mpc
df['Dcm'] = df['DL(MPC)']*(3.086e+24)

df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,6))

nan_mask = df['L2-10intr_flag'].isnull()

im1 = ax1.scatter(x=df['das'], y=df['L2-10obs'], c='royalblue', label='Observed 2-10 keV luminosity')
im1 = ax1.scatter(x=df.loc[nan_mask, 'das'], y=df.loc[nan_mask, 'L2-10intr'], c='tab:orange', label='Intrinsic 2-10 keV luminosity')

im1 = ax1.scatter(x=df.loc[~nan_mask,'das'], y=df.loc[~nan_mask, 'L2-10intr'], c='tab:orange', marker='^', label='Intrinsic 2-10 keV luminosity (lower limit)')

nan_mask = df['L10-24intr_flag'].isnull()

im2 = ax2.scatter(x=df['das'], y=df['L10-24obs'], c='royalblue', label='Observed 10-24 keV luminosity')
im2 = ax2.scatter(x=df.loc[nan_mask, 'das'], y=df.loc[nan_mask, 'L10-24intr'], c='tab:orange', label='Intrinsic 10-24 keV luminosity')

im2 = ax2.scatter(x=df.loc[~nan_mask,'das'], y=df.loc[~nan_mask, 'L10-24intr'], c='tab:orange', marker='^', label='Intrinsic 10-24 keV luminosity (lower limit)')

ax1.set_ylim(40.5,44.5)
ax2.set_ylim(40.5,44.5)
    
ax1.set(xlabel="Angular Separation (\'\')", ylabel="L$_{\mathrm{2-10\,keV}}$")
ax2.set(xlabel="Angular Separation (\'\')", ylabel="L$_{\mathrm{10-24\,keV}}$")

ax1.legend()
ax2.legend()

ax1.set_xscale('log')
ax2.set_xscale('log')

plt.tight_layout()
plt.savefig('luminosities_v_angsep.png', dpi=150)
```

```python
df['F2-10obs'] = (10**df['L2-10obs'])/(4*np.pi*(df['Dcm']**2))
df['F2-10intr'] = (10**df['L2-10intr'])/(4*np.pi*(df['Dcm']**2))
df['F10-24obs'] = (10**df['L10-24obs'])/(4*np.pi*(df['Dcm']**2))
df['F10-24intr'] = (10**df['L10-24intr'])/(4*np.pi*(df['Dcm']**2))

df['F10-24obs_flag'] = df['L10-24obs_flag']
df['F2-10intr_flag'] = df['L2-10intr_flag']
df['F10-24intr_flag'] = df['L10-24intr_flag']

df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,6))

nan_mask = df['F2-10intr_flag'].isnull()

im1 = ax1.scatter(x=df['das'], y=df['F2-10obs'], c='royalblue', label='Observed 2-10 keV luminosity')
im1 = ax1.scatter(x=df.loc[nan_mask, 'das'], y=df.loc[nan_mask, 'F2-10intr'], c='tab:orange', label='Intrinsic 2-10 keV luminosity')

im1 = ax1.scatter(x=df.loc[~nan_mask,'das'], y=df.loc[~nan_mask, 'F2-10intr'], c='tab:orange', marker='^', label='Intrinsic 2-10 keV luminosity (lower limit)')

nan_mask = df['F10-24intr_flag'].isnull()

im2 = ax2.scatter(x=df['das'], y=df['F10-24obs'], c='royalblue', label='Observed 10-24 keV luminosity')
im2 = ax2.scatter(x=df.loc[nan_mask, 'das'], y=df.loc[nan_mask, 'F10-24intr'], c='tab:orange', label='Intrinsic 10-24 keV luminosity')

im2 = ax2.scatter(x=df.loc[~nan_mask,'das'], y=df.loc[~nan_mask, 'F10-24intr'], c='tab:orange', marker='^', label='Intrinsic 10-24 keV luminosity (lower limit)')

ax1.set_ylim(3e-14,2e-10)
ax2.set_ylim(3e-14,2e-10)
    
ax1.set(xlabel="Angular Separation (\'\')", ylabel="F$_{\mathrm{2-10\,keV}}$")
ax2.set(xlabel="Angular Separation (\'\')", ylabel="F$_{\mathrm{10-24\,keV}}$")

ax1.legend()
ax2.legend()

ax1.set_xscale('log')
ax2.set_xscale('log')
ax1.set_yscale('log')
ax2.set_yscale('log')

plt.tight_layout()
plt.savefig('fluxes_v_angsep.png', dpi=150)
```
